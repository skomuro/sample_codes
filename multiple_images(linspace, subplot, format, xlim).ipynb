```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
taus2 = np.arange(0.8, 2.0 , 0.1)
```

```python
plt.figure(figsize=(45, 30))
for k in range(len(taus2)):
    tau = taus2[k]
    xs = np.linspace(-15, 15, 1000)
    ys = np.sin(tau * xs * np.pi)
    plt.subplot(3, 4, k+1)
    plt.plot(facecolor = "white")
    plt.plot(xs, ys)
    plt.xlim(-15, 15)
    plt.ylim(-1, 1)
    plt.title(r"$\tau={:.1f} x \pi$".format(tau))
    plt.xlabel("Final Epidemic Size")
    plt.ylabel("Frequency")
```
